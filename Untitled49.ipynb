```python
import sys

try:
    get_ipython  # noqa
    IN_IPY = True
except Exception:
    IN_IPY = False

if IN_IPY:
    try:
        # Use Jupyter-friendly installer; it's okay if already installed
        %pip -q install pandas numpy scikit-learn nltk beautifulsoup4 html5lib vaderSentiment pyyaml joblib matplotlib seaborn wordcloud gradio
    except Exception as e:
        print("[WARN] Could not auto-install packages:", e)

# %% [markdown]
# ## 1) Imports & Config

# %%
from __future__ import annotations
import sqlite3, hashlib, json, yaml, joblib, re, string, warnings
from pathlib import Path
from typing import List, Optional, Tuple, Dict

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix

import nltk
from bs4 import BeautifulSoup
from nltk.stem import WordNetLemmatizer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

warnings.filterwarnings("ignore")

# --- Set your paths here ---
IN_SQLITE = Path(r"C:\Users\NXTWAVE\Downloads\Review Sentiment Analyzer\archive\database.sqlite")
IN_CSV    = Path(r"C:\Users\NXTWAVE\Downloads\Review Sentiment Analyzer\archive\Reviews.csv")
IN_HASHES = Path(r"C:\Users\NXTWAVE\Downloads\Review Sentiment Analyzer\archive\hashes.txt")
OUT_DIR   = Path(r"C:\Users\NXTWAVE\Downloads\Review Sentiment Analyzer")

# Candidate column names found in common datasets
TEXT_CANDIDATES   = ["reviewText","Text","text","content","review"]
RATING_CANDIDATES = ["rating","Rating","Score","stars","star_rating"]
PID_CANDIDATES    = ["productID","ProductId","asin","product_id","item_id"]
CAT_CANDIDATES    = ["category","productCategory","Category"]
USER_CANDIDATES   = ["userID","UserId","profileName","reviewerID"]
SUM_CANDIDATES    = ["summary","Summary","title"]
TIME_CANDIDATES   = ["time","Time","timestamp","unixReviewTime","date"]

lemmatizer = WordNetLemmatizer()
URL_RE = re.compile(r"https?://\S+|www\.\S+")
PUNCT_TABLE = str.maketrans('', '', string.punctuation)

# Download minimal NLTK data (silent)
try:
    nltk.download('wordnet', quiet=True)
except Exception:
    pass

# %% [markdown]
# ## 2) Utilities

# %%
def ensure_out():
    OUT_DIR.mkdir(parents=True, exist_ok=True)


def pick_first(df: pd.DataFrame, cands: List[str]) -> Optional[str]:
    for c in cands:
        if c in df.columns:
            return c
    return None


def normalize_cols(df: pd.DataFrame) -> pd.DataFrame:
    cols = list(df.columns)
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]
    tcol = pick_first(df, TEXT_CANDIDATES)
    if not tcol:
        raise ValueError(f"No text column found in {cols}")
    rcol = pick_first(df, RATING_CANDIDATES)
    pcol = pick_first(df, PID_CANDIDATES)
    ccol = pick_first(df, CAT_CANDIDATES)
    ucol = pick_first(df, USER_CANDIDATES)
    scol = pick_first(df, SUM_CANDIDATES)
    tcol2 = pick_first(df, TIME_CANDIDATES)
    keep = {tcol:"reviewText"}
    if rcol: keep[rcol] = "rating"
    if pcol: keep[pcol] = "productID"
    if ccol: keep[ccol] = "productCategory"
    if ucol: keep[ucol] = "userID"
    if scol: keep[scol] = "summary"
    if tcol2: keep[tcol2] = "time"
    out = df[list(keep.keys())].rename(columns=keep)
    return out


def load_sqlite(path: Path) -> List[pd.DataFrame]:
    res = []
    if not path.exists():
        return res
    con = sqlite3.connect(str(path))
    try:
        tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", con)
        for t in tables['name']:
            try:
                df = pd.read_sql(f"SELECT * FROM [{t}]", con)
                if any(c in df.columns for c in TEXT_CANDIDATES):
                    dfN = normalize_cols(df)
                    dfN["_source_table"] = t
                    res.append(dfN)
            except Exception:
                continue
    finally:
        con.close()
    return res


def load_csv(path: Path) -> Optional[pd.DataFrame]:
    if not path.exists():
        return None
    df = pd.read_csv(path)
    try:
        out = normalize_cols(df)
        out["_source_table"] = "CSV"
        return out
    except Exception:
        return None


def clean_text(s: str) -> str:
    if not isinstance(s, str):
        return ""
    t = BeautifulSoup(s, "html5lib").get_text(" ")
    t = URL_RE.sub(" ", t)
    t = t.lower().translate(PUNCT_TABLE)
    t = re.sub(r"[^a-z0-9\s]", " ", t)
    t = re.sub(r"\s+", " ", t).strip()
    return t


def sha1_of_text(s: str) -> str:
    return hashlib.sha1((s or "").encode('utf-8')).hexdigest()


def apply_hash_dedup(df: pd.DataFrame, hashes_path: Path) -> pd.DataFrame:
    df = df.copy()
    df['cleanText'] = df['reviewText'].astype(str).map(clean_text)
    df['hash'] = df['cleanText'].map(sha1_of_text)

    seen = set()
    if hashes_path.exists():
        with open(hashes_path, 'r', encoding='utf-8', errors='ignore') as f:
            seen = {line.strip() for line in f if line.strip()}

    before = len(df)
    df = df[~df['hash'].isin(seen)].drop_duplicates(subset=['hash']).reset_index(drop=True)
    after = len(df)
    print(f"[DEDUP] removed {before - after} rows using hashes.txt + internal dedup")
    return df


def update_hashes(df: pd.DataFrame, hashes_path: Path):
    if 'hash' not in df.columns:
        return
    existing = set()
    if hashes_path.exists():
        with open(hashes_path, 'r', encoding='utf-8', errors='ignore') as f:
            existing = {line.strip() for line in f if line.strip()}
    new_hashes = [h for h in df['hash'].tolist() if h and h not in existing]
    if new_hashes:
        with open(hashes_path, 'a', encoding='utf-8') as f:
            for h in new_hashes:
                f.write(h + "\n")
        print(f"[DEDUP] appended {len(new_hashes)} new hashes to hashes.txt")

# %% [markdown]
# ## 3) Sentiment & Model

# %%
def rule_score(text: str) -> float:
    vs = SentimentIntensityAnalyzer()
    return float(vs.polarity_scores(text).get('compound', 0.0))


def score_to_label(score: float, pos=0.2, neg=-0.2) -> str:
    return 'positive' if score>=pos else ('negative' if score<=neg else 'neutral')


def prepare_labels(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    if 'rating' in out.columns:
        out['rating'] = pd.to_numeric(out['rating'], errors='coerce')
        out['label'] = out['rating'].apply(lambda r: 'positive' if r>=4 else ('negative' if r<=2 else 'neutral'))
    else:
        out['label'] = out['cleanText'].apply(lambda t: score_to_label(rule_score(t)))
    return out


def train_model(df: pd.DataFrame, seed=42):
    y = df['label']
    X = df['cleanText']
    if (y=="neutral").sum() < 50:
        keep = y!="neutral"
        X, y = X[keep], y[keep]
    Xtr, Xva, ytr, yva = train_test_split(X, y, test_size=0.2, random_state=seed, stratify=y)
    pipe = Pipeline([
        ("tfidf", TfidfVectorizer(max_features=50000, ngram_range=(1,2))),
        ("clf", LogisticRegression(C=4.0, max_iter=250, n_jobs=-1))
    ])
    pipe.fit(Xtr, ytr)
    yhat = pipe.predict(Xva)
    rep = classification_report(yva, yhat, output_dict=True)
    cm  = confusion_matrix(yva, yhat, labels=sorted(y.unique()))
    eval_df = pd.DataFrame({"text":Xva, "label_true":yva, "label_pred":yhat})
    return pipe, rep, cm, eval_df

# %% [markdown]
# ## 4) Load, Clean, Dedup, Label

# %%
ensure_out()

frames: List[pd.DataFrame] = []
frames += load_sqlite(IN_SQLITE)
csv_df = load_csv(IN_CSV)
if csv_df is not None:
    frames.append(csv_df)

if not frames:
    raise SystemExit("No usable tables or CSV found  check your paths.")

df = pd.concat(frames, ignore_index=True, sort=False)
print(f"[LOAD] Combined rows: {len(df)} | cols: {df.columns.tolist()}")

# Clean & dedup using hashes.txt
df = df.rename(columns={c:c for c in df.columns})
df['reviewText'] = df['reviewText'].astype(str)
df = apply_hash_dedup(df, IN_HASHES)

# Labeling
df = prepare_labels(df)
print(df.head(3))

# %% [markdown]
# ## 5) Train & Evaluate

# %%
model, report, cm, eval_df = train_model(df)
print(pd.DataFrame(report).round(3))
print("\nConfusion Matrix (labels sorted):\n", cm)

# Save evaluation predictions
(eval_df).to_csv(OUT_DIR / 'eval_predictions_valid.csv', index=False)

# %% [markdown]
# ## 6) Visuals (Distribution, Heatmap, Wordclouds)

# %%
# Distribution
plt.figure()
sns.countplot(x='label', data=df, order=['negative','neutral','positive'])
plt.title('Sentiment distribution')
plt.tight_layout()
plt.savefig(OUT_DIR / 'viz_sentiment_distribution.png', dpi=160)
plt.show()

# Heatmap rating x sentiment (if rating exists)
if 'rating' in df.columns:
    tmp = df.copy()
    tmp['rating'] = pd.to_numeric(tmp['rating'], errors='coerce')
    tmp = tmp.dropna(subset=['rating','label'])
    pivot = pd.crosstab(tmp['rating'], tmp['label'])
    if not pivot.empty:
        plt.figure()
        sns.heatmap(pivot, annot=True, fmt='d')
        plt.title('Rating  Sentiment')
        plt.tight_layout()
        plt.savefig(OUT_DIR / 'viz_rating_x_sentiment_heatmap.png', dpi=160)
        plt.show()

# Wordclouds
texts = {
    'positive': ' '.join(df[df['label']=='positive']['reviewText'].astype(str).tolist())[:3_000_000],
    'negative': ' '.join(df[df['label']=='negative']['reviewText'].astype(str).tolist())[:3_000_000],
}
for lab, txt in texts.items():
    if txt:
        wc = WordCloud(width=1200, height=600, background_color='white').generate(txt)
        wc.to_file(str(OUT_DIR / f'viz_wordcloud_{lab}.png'))
        plt.figure(figsize=(10,5))
        plt.imshow(wc)
        plt.axis('off')
        plt.title(f'Wordcloud  {lab}')
        plt.show()

# %% [markdown]
# ## 7) Insights JSON & Hidden-Issue Products

# %%
insights: Dict[str, object] = {}
insights['sentiment_distribution'] = df['label'].value_counts().to_dict()

hidden = pd.DataFrame()
if 'rating' in df.columns and 'productID' in df.columns:
    tmp = df.copy()
    tmp['rating'] = pd.to_numeric(tmp['rating'], errors='coerce')
    hidden = tmp.groupby('productID').agg(
        avg_rating=('rating','mean'),
        n_reviews=('reviewText','count'),
        neg_frac=('label', lambda s: float((s=='negative').mean()))
    ).reset_index()
    hidden = hidden[(hidden['avg_rating']>=4) & (hidden['neg_frac']>=0.15) & (hidden['n_reviews']>=10)]
    hidden = hidden.sort_values(['neg_frac','n_reviews'], ascending=[False,False]).head(20)

insights['top_products_hidden_issues'] = hidden.to_dict(orient='records') if not hidden.empty else []

with open(OUT_DIR / 'insights.json', 'w', encoding='utf-8') as f:
    json.dump(insights, f, ensure_ascii=False, indent=2)

print("Insights saved to:", OUT_DIR / 'insights.json')

# %% [markdown]
# ## 8) Save Artifacts (H5 / PKL / YAML) + Update hashes

# %%
# HDF5
(df.drop(columns=['cleanText'], errors='ignore')).to_hdf(OUT_DIR / 'processed_reviews.h5', key='reviews', mode='w')

# PKL model
joblib.dump(model, OUT_DIR / 'model_sentiment.pkl')

# YAML metadata + metrics
meta = {
    'inputs': {
        'sqlite': str(IN_SQLITE),
        'csv': str(IN_CSV),
        'hashes': str(IN_HASHES),
    },
    'schema': list(df.columns),
    'counts': {
        'rows': int(len(df)),
        'labels': df['label'].value_counts().to_dict(),
    },
    'model': 'TFIDF+LogisticRegression',
    'metrics': report,
}
with open(OUT_DIR / 'build_metadata.yaml', 'w', encoding='utf-8') as f:
    yaml.safe_dump(meta, f, sort_keys=False)

# Update hashes.txt
if not IN_HASHES.parent.exists():
    IN_HASHES.parent.mkdir(parents=True, exist_ok=True)
update_hashes(df, IN_HASHES)

print("\n[OK] Artifacts written:")
print(" ", OUT_DIR / 'processed_reviews.h5')
print(" ", OUT_DIR / 'model_sentiment.pkl')
print(" ", OUT_DIR / 'build_metadata.yaml')
print(" ", OUT_DIR / 'insights.json')
print(" ", OUT_DIR / 'eval_predictions_valid.csv')

# %% [markdown]
# ## 9) (Optional) Gradio UI  Run locally inside Jupyter
# - After running this cell, open the public/local URL it prints.

# %%
import gradio as gr

def _predict_one(text: str, model_path=str(OUT_DIR / 'model_sentiment.pkl')) -> Tuple[str, float]:
    if not text.strip():
        return "", 0.0
    try:
        pipe = joblib.load(model_path)
        lab = pipe.predict([text])[0]
        try:
            proba = pipe.predict_proba([text])
            conf = float(np.max(proba))
        except Exception:
            conf = 0.0
        return lab, conf
    except Exception:
        # Fallback to rule-based
        comp = rule_score(text)
        return score_to_label(comp), float(abs(comp))

with gr.Blocks(title="ReviewSentiment  Notebook UI") as demo:
    gr.Markdown("#  ReviewSentiment  Notebook UI\nEnter a review and get instant sentiment.")
    inp = gr.Textbox(label="Review text", lines=4)
    btn = gr.Button("Predict")
    out1 = gr.Textbox(label="Sentiment")
    out2 = gr.Number(label="Confidence / Score")
    btn.click(_predict_one, inputs=inp, outputs=[out1, out2])

# Uncomment to launch from within notebook
# demo.launch(server_name="0.0.0.0", server_port=7860, show_error=True)
```
